```python
# Element 접근하는 세 메소드
def get_element(xpath):
    try:
        time.sleep(0.2)
        return WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, xpath)))
    except StaleElementReferenceException:
        time.sleep(0.2)
        return get_element(xpath)

def get_element_text(xpath):
    try:
        time.sleep(0.2)
        txt=driver.find_element(By.XPATH,xpath).text
        return txt
    except:
        txt=""
        return txt

def get_element_click(xpath):
    try:
        time.sleep(0.2)
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, xpath))).click()
    except StaleElementReferenceException:
        time.sleep(0.2)
        get_element_click(xpath)
```

```python
errors=[]
videos=[]
idx=0
total_idx=len(url_list)

# 각 URL마다 xpath0 xpath1 xpath2 xpath3 클릭
for url in url_list:
    idx+=1
        
    try:    
        print(f"({idx}/{total_idx}) 진행중 : {url}")

        # URL에 접속
        driver.get(f"https://music.youtube.com/search?q={url} song")
        time.sleep(0.5)

        # 음원 없는 경우 처리
        if get_element_text(xpathx)=="동영상":
            videos.append(url)
            print("동영상 추가 : "+url)
            continue

        # xpath0 클릭
        get_element_click(xpath0)

        # 페이지 이동 확인 후 xpath1 우클릭
        while get_element_text(xpath1) =="노래":
            True
        ActionChains(driver).context_click(get_element(xpath1)).perform()

        # xpath2 클릭
        get_element_click(xpath2)

        # xpath3 클릭
        get_element_click(xpath3)
    
    except Exception as e:
        errors.append(url)
        print("{} : {}".format(traceback.format_exc(), url))


# 노음원 출력
with open('동영상.txt', 'w') as f:
    for item in videos:
        f.write("%s\n" % item)
        
# 오류 출력
with open('오류.txt', 'w') as f:
    for item in errors:
        f.write("%s\n" % item)
```
